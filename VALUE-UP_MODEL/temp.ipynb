```python
# Data\ValueUp
# 1.marketcap
# 2.fundamentals
# 3.dividends

# 폴더 이름
# 1. marketcap
 # 칼럼 명 : stockcode,avg_marcap,avg_amount,Name,Market,Stocks
# 2. fundamentals
 # 날짜,BPS,PER,PBR,EPS,DIV,DPS,Year,stockcode,Name,Market,년도
# 3. dividends
 # stockcode,Name,Market,Stocks,avg_marcap,avg_amount,2018_배당,2018_자사주,2019_배당,2019_자사주,최근2년_활동

# 라이브러리 호출
import pandas as pd
import numpy as np
import os

marketcap = pd.read_csv('../Data/ValueUp/1.marketcap/cap_2018.csv')
marketcap

year = list(range(2018, 2025))

# 메타 데이터 추출을 위해 모든 데이터 병합

marketcap_list = []
for year in year:
    marketcap = pd.read_csv('../Data/ValueUp/1.marketcap/cap_{}.csv'.format(year))
    marketcap_list.append(marketcap)

marketcap_list

# 데이터 병합
marketcap_all = pd.concat(marketcap_list, axis=0)

marketcap_all
```

```python
# 시가총액 기준으로 정렬하고 종목코드 중복 제거
stock_code = marketcap_all.sort_values(['avg_marcap', 'avg_amount'], ascending=[False, False])

# stocode 별로 하나 씩만 남기기
stock_code = stock_code.drop_duplicates(subset=['stockcode'])

# 종목코드 출력
stock_code

# 인덱스 리셋
stock_code = stock_code.reset_index(drop=True)

# 종목코드 출력
stock_code.to_csv('../Data/metadata.csv', index=False)
```
